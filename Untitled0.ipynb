```python
# Resizing using pillow library
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')
img_resize = img.resize((200,200))

```

```python
img_resize.save("dog_resized_img.jpg")
```

```python
# display resized image from numply array

img_plot = plt.imshow(img_resize)
plt.show()
```

```python
#converting rgb to greyscale

import cv2

img = cv2.imread("/content/dog.jpg")

greyscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(greyscale_img)
```

```python
print(greyscale_img.shape)
```

```python
img_plot = plt.imshow(greyscale_img)
plt.show()
```

```python
cv2.imwrite('greyscale_img.jpg', greyscale_img)
```
